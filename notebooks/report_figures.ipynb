```python
import os
import cv2
dataPath = "F:\\datasets\\eirt_output\\batch01"
depth_dir = os.path.join(dataPath, "depth")
rgb_dir = os.path.join(dataPath, "rgb")
mask_dir = os.path.join(dataPath, "mask_raw")
img_name = "Image0016"

depth_img = cv2.cvtColor(cv2.imread(os.path.join(depth_dir, f"{img_name}.png"), cv2.IMREAD_UNCHANGED), cv2.COLOR_BGR2GRAY)
rgb_img = cv2.imread(os.path.join(rgb_dir, f"{img_name}.png"))
mask_img = cv2.imread(os.path.join(mask_dir, f"{img_name}.jpg"), cv2.IMREAD_GRAYSCALE)
print(depth_img.shape)

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Depth Image")
plt.imshow(depth_img)
plt.subplot(1, 3, 2)
plt.title("RGB Image")
plt.imshow(cv2.cvtColor(rgb_img, cv2.COLOR_BGR2RGB))
plt.subplot(1, 3, 3)
plt.title("Mask Image")
plt.imshow(mask_img, cmap='gray')
plt.show()

# Save depth_img with viridis colormap
save_path = r"D:\UNI\Semester 9\eirt\report"
os.makedirs(save_path, exist_ok=True)

# Apply viridis colormap and save
plt.figure(figsize=(8, 6))
plt.imshow(depth_img, cmap='viridis')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig(os.path.join(save_path, f"{img_name}_depth_viridis.png"), bbox_inches='tight', pad_inches=0, dpi=150)
plt.close()
print(f"Depth image with viridis colormap saved to {os.path.join(save_path, f'{img_name}_depth_viridis.png')}")
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(depth_img.flatten(), bins=255, alpha=0.7)
plt.xlabel('Pixel Values')
plt.ylabel('Frequency (log scale)')
plt.title('Histogram of Pixel Values')
plt.yscale('log')
plt.show()
```

```python
#depth histogram equalization
equalized_depth = cv2.equalizeHist(depth_img)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title("Original Depth Image")
plt.imshow(depth_img, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Equalized Depth Image")
plt.imshow(equalized_depth, cmap='gray')
plt.show()
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title("Original Depth Image")
plt.imshow(depth_img)
plt.subplot(1, 2, 2)
plt.title("Equalized Depth Image")
plt.imshow(equalized_depth)
plt.show()
```
